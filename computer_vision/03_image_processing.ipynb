```python
import numpy as np
import argparse
import imutils
import cv2
from PIL import Image

FILE_NAME = 'draw.png'
```

```python
ap = argparse.ArgumentParser()
ap.add_argument("-i", "--image", help="./images/trex.png", default="./images/trex.png")
ap.add_argument("-f")
args = vars(ap.parse_args())

image = cv2.imread(args['image'])
Image.open('./images/trex.png')
```

```python
M = np.float32([[1, 0, 25], [0, 1, 50]])
shifted = imutils.translate(image, 25, 50)

cv2.imwrite(FILE_NAME, shifted)
Image.open(FILE_NAME)
```

```python
shifted = imutils.translate(image, -50, -70)

cv2.imwrite(FILE_NAME, shifted)
Image.open(FILE_NAME)
```

```python
rotated = imutils.rotate(image, 180)

cv2.imwrite(FILE_NAME, rotated)
Image.open(FILE_NAME)
```

```python
ratio = 900 / image.shape[1]
dim = (900, int(image.shape[0] * ratio))

resized = imutils.resize(image, height=100, width=88)
cv2.imwrite(FILE_NAME, resized)
Image.open(FILE_NAME)
```

```python
flipped = cv2.flip(image, -1)

cv2.imwrite(FILE_NAME, flipped)
Image.open(FILE_NAME)
```

```python
cropped = image[30:120, 240:-1]

cv2.imwrite(FILE_NAME, cropped)
Image.open(FILE_NAME)
```

```python
import argparse
import cv2
```

```python
ap = argparse.ArgumentParser()

ap.add_argument("-i", "--image", required = False, help="./images/trex.png", default='./images/trex.png')
ap.add_argument('-f')
args = vars(ap.parse_args())
```

```python
image = cv2.imread(args["image"])
print("width: {}".format(image.shape[1]))
print("height: {}".format(image.shape[0]))
print("channels: {}".format(image.shape[2]))

from PIL import Image
Image.open('./images/trex.png')
```

```python
(b, g, r) = image[0, 0]
print("Pixel at (0,0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))

image[0, 0] = (0, 0, 255)
(b, g, r) = image[0, 0]
print("Pixel at (0,0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
```

```python
corner = image[0:100, 200:-1]
image[0:100, 0:100] = (0, 255, 0)

cv2.imwrite("draw.png", image)
Image.open('draw.png')
```
